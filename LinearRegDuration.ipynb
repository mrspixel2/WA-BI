```python
X = TMP[[ 'idResource', 'idProject',

         'comment',
        'idRole', 'dailyCost',
       'realCost', 'leftCost', 'notPlannedWork']]
y = TMP['realWork']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif      
from sklearn.feature_selection import f_regression  
#Score function:
#For regression: f_regression, mutual_info_regression
#For classification: chi2, f_classif, mutual_info_classif

kbest = SelectKBest(score_func=f_regression, k=5)
kbest.fit(X_train, y_train)
print("Sélection de variables :", kbest.get_support())
print("Scores de variables :", kbest.scores_)
print("Variables sélectionnées:", list(X.columns[kbest.get_support()]))
print("Variables supprimées :", list(X.columns[~kbest.get_support()]))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, median_absolute_error, r2_score

X_train = kbest.transform(X_train)
X_test = kbest.transform(X_test)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print('test_score = ',model.score(X_test,y_test)) 
print('R2 = ',r2_score(y_test, y_pred))
print('MAE = ',mean_squared_error(y_test, y_pred))
print('RMSE = ', np.sqrt(mean_squared_error(y_test, y_pred)))
print('MAE = ', mean_absolute_error(y_test, y_pred))
print('Meadian = ', median_absolute_error(y_test, y_pred))
```

```python
import flask

import pickle

# Use pickle to load in the pre-trained model.

with open(f'model/duration.pkl', 'rb') as f:

    model = pickle.load(f)

app = flask.Flask(__name__, template_folder='templates')


@app.route('/duration', methods=['GET', 'POST'])

def duration():

    if flask.request.method == 'GET':

        return(flask.render_template('main.html'))

    if flask.request.method == 'POST': 

        idResource = flask.request.form['idResource']

        idProject = flask.request.form['idProject']

        realCost = flask.request.form['realCost']
        
        idRole = flask.request.form['idRole']

        dailyCost = flask.request.form['dailyCost']

        input_variables = pd.DataFrame([[idResource, idProject, realCost,idRole,dailyCost]],

                                       columns=['idResource', 'idProject', 'realCost', 'idRole', 'dailyCost'],

                                       dtype=float)

        prediction = model.predict(input_variables)[0]

        return flask.render_template('main.html',

                                     original_input={'idResource':idResource,

                                                     'idProject':idProject,

                                                     'realCost':realCost,

                                                     'idRole':idRole,

                                                     'dailyCost':dailyCost},

                                     result=prediction,)

from werkzeug.serving import run_simple
run_simple('localhost', 8002, app)
```
